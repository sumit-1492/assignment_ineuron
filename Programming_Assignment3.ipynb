# 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?





```python
def pnz():
    try:
        num = float(input("Enter any number as per your wish: "))
        if num>0:
            print("It is a positive number : ",num)
        elif num<0:
            print("It is a negative number : ",num)
        elif num == 0:
            print("The value of number is zero ")
    except Exception as e:
        print("Error message","\n",e)
pnz()
```

# 2.	Write a Python Program to Check if a Number is Odd or Even?

```python
def oe():
    try:
        num = float(input("Enter any number as per your wish: "))
        if num%2 == 0:
            print("It is a even number : ",num)
        else :
            print("It is a odd number : ",num)
    except Exception as e:
        print("Error message","\n",e)
oe()
```

# 3.	Write a Python Program to Check Leap Year?

```python
year = int(input("Enter the year as per your wish: "))
if year%4 == 0:
    print("The given year is a leap year")
```

# 4.	Write a Python Program to Check Prime Number?

```python
def cpn():
    num = int(input("Enter the number as per your wish: "))
    for i in range(2,num):
        if num%i == 0:
            print("It is not a prime number: ",num)
            break;
        else:
            print("It is a prime number")
            break;
cpn()
```
